```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
sns.countplot(df,x='class',)
plt.title('number of pesenger in each class')
plt.xlabel('class')
plt.ylabel('numberof passengers')
plt.show()
```

```python
sns.histplot(df,x='age')
plt.title('age distribution of passenger')
plt.xlabel('age')
plt.ylabel('number of passenger')
plt.show()
```

```python
sns.histplot(df,x='age',kde=True, color ='red') #kde define wave
plt.title('age distribution of passenger')
plt.xlabel('age')
plt.ylabel('number of passenger')
plt.show()
```

```python
sns.lineplot(df,x="class",y="fare",linestyle='dashed',marker='o',color='green')
plt.title('fare of each class')
plt.xlabel('class')
plt.ylabel('fare')
plt.show()
```
